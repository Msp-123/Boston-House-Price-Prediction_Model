```python
# Spliting data to have 'dataset_x' and 'dataset_y'
dataset_x = data.drop(['PRICE'], axis=1).to_numpy()
dataset_y = data['PRICE'].to_numpy()
```

```python
dataset_x.shape
```

```python
dataset_y.shape
```

```python
# Training and test datasets
from sklearn.model_selection import train_test_split
training_dataset_x, test_dataset_x, training_dataset_y, test_dataset_y = train_test_split(dataset_x, dataset_y, test_size = 0.2, random_state = 3)
```

```python
training_dataset_x
```

```python
training_dataset_y
```

```python
# Scaling of columns of dataset_x
# After checking the histograms of the variables StandardScaler did not seem appropriate
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(training_dataset_x)
scaled_training_dataset_x = mms.transform(training_dataset_x)
scaled_test_dataset_x = mms.transform(test_dataset_x)
```

```python
scaled_training_dataset_x
```

## MODELS

### Linear Regression Model

```python
# Importing library for Linear Regression and Creating a Linear Regressor
from sklearn.linear_model import LinearRegression
linear_regression_model = LinearRegression()
```

```python
# Cross_val_score of linear regression_model
import sklearn.metrics
from sklearn.model_selection import cross_val_score
linear_reg_eval=cross_val_score(estimator = linear_regression_model,
                                X=scaled_training_dataset_x,y=training_dataset_y, scoring='r2',cv=5 )
```

```python
linear_reg_eval
```

### Random Forest Regressor

```python
# Import Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

#Create a Random Forest Regressor
reg = RandomForestRegressor()
```

```python
# Cross_val_score of linear regression_model
import sklearn.metrics
from sklearn.model_selection import cross_val_score
reg_eval=cross_val_score(estimator = reg,
                                X=scaled_training_dataset_x,y=training_dataset_y, scoring='r2',cv=5 )
```

```python
reg_eval
```

### XGBoost Regressor


```python
# Import XGBoost Regressor
from xgboost import XGBRegressor

# Create a XGBBoost Regressor
xgb = XGBRegressor()
```

```python
# Cross_val_score of linear regression_model
import sklearn.metrics
from sklearn.model_selection import cross_val_score
xgb_eval=cross_val_score(estimator = xgb,
                                X=scaled_training_dataset_x,y=training_dataset_y, scoring='r2',cv=5 )
```

```python
xgb_eval
```

### Parameter Optimization with GridSearch Method

```python
search_space = {
    'nthread':[3,4, 5],  # when using hyperthread, xgboost may become slower
    'objective':['reg:linear'],
    'learning_rate':[0.03, 0.05, 0.1, 0.3, 0.7], # step size at each iteration
    'max_depth':[5, 6, 7], # the max depth per tree. A deeper tree might increase performance and also complexity and chance to overfit
    'min_child_weight': [4], 
    'random_state': [8],
    'n_estimators':[100, 130, 160] # the number of tress in our ensemble
}
```

```python
from sklearn.model_selection import GridSearchCV

GS = GridSearchCV(estimator = xgb, param_grid = search_space, scoring = ['r2', 'neg_root_mean_squared_error'],
                  refit='r2', cv=5 )
# All scorer objects follow the convention that higher return values are better than lower return values.
# Hence, 'neg_mse' is used instead of 'mse'
```

```python
GS.fit(training_dataset_x, training_dataset_y)
```

```python
print(GS.best_params_)
```

```python
print(GS.best_score_)
```
